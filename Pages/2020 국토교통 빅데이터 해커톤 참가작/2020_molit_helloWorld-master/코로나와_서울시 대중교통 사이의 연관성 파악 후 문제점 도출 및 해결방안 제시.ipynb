3. 집단 감염 case 일자 전후로 대중교통 이용량을 비교하여, 실제로 COVID19가 대중교통 이용량 감소에 영향을 미친다는 사실을 확인하였습니다.

5. 실제로 서울교통공사 지하철의 혼잡도를 분석한 결과, 모든 호선에서 100% 이상의 혼잡도가 나타났습니다.
6. 이러한 혼잡도 문제를 해결하고, 국민들의 안전한 이동을 도모하기 위해 대체 교통수단 공공앱 '비상구'를 제안합니다. 해당 앱은 대중교통 실시간 혼잡도 안내, 혼잡한 지역의 대체 교통수단 제시, 개인 맞춤형 경로 제공 등의 기능을 갖추고 있습니다.

## 1. COVID19와 서울시 대중교통의 상관관계 파악

### 서울시 구별 확진자 수와 구별 대중교통의 상관관계 분석

```python
# 구별 확진자 수 파악
# 데이터 불러오기 (PatientInfo)
covid = pd.read_csv('../data/PatientInfo.csv',encoding = 'utf-8')
# 서울시만 추출
covid_seoul = covid.loc[covid['province']=='서울',:]
# 구별 확진자수 세기
covid_seoul = covid_seoul.groupby('city').count()
# 필요한 열만 추출
covid_seoul = covid_seoul[['patient_id']]
```

```python
#구별 대중교통 이용량 확인
#데이터 불러오기
#교통카드 빅데이터 중 시간별 이용인원
gu_st_total = pd.read_csv('../data/DM_TMBY_USESTF_SMY.dat',
                  sep='|',
                  encoding='utf-8',
                  header=None)
# 열 이름 설정
gu_st_total.columns = ['Operation_Date','SiDo_Code','SiGunGu_Code','Time','Number_of_users']
# ['시도코드'] 열 - 물결로 나타나 있는 값 > 문자열 0으로 바꾸기
gu_st_total.loc[(gu_st_total['SiDo_Code'] == '~'), 'SiDo_Code'] = '0'
# ['시군구코드'] 열 - 물결로 나타나 있는 값 > 문자열 0으로 바꾸기
gu_st_total.loc[(gu_st_total['SiGunGu_Code'] == '~'), 'SiGunGu_Code'] = '0'
# ['시간대'] 열 -  물결로 나타나 있는 값 > 문자열 0으로 바꾸기
gu_st_total.loc[(gu_st_total['Time'] == '~'), 'Time'] = '0'
# 나중에 계산 편하게 하기 위해 object를 int로 바꿔놓기
gu_st_total.dtypes
gu_st_total['SiDo_Code'] = pd.to_numeric(gu_st_total['SiDo_Code'])
gu_st_total['SiGunGu_Code'] = pd.to_numeric(gu_st_total['SiGunGu_Code'])
gu_st_total['Time'] = pd.to_numeric(gu_st_total['Time'])
# 서울시 데이터만 사용할것이므로 시도코드 11만 출력
# new data frame name : gu_st_seoul
gu_st_seoul = gu_st_total[gu_st_total['SiDo_Code'] == 11]
# 이후에 사용하기 위해 지금까지의 데이터 저장 
gu_st_seoul_m = gu_st_seoul
gu_st_seoul_d = gu_st_seoul
gu_st_seoul_y = gu_st_seoul
# 필요없는 열 제외
gu_st_seoul = gu_st_seoul[['SiGunGu_Code','Number_of_users']]
#구별 이용량 확인
gu_st_seoul_pivot = gu_st_seoul.pivot_table(index='SiGunGu_Code', aggfunc=sum)
# 시군구코드에 맞춰서 행이름을 바꾸기 위해 
# 지역 코드 확인하여 서울특별시만 추출
# DD_AREA.dat 파일 로드
# df명 area
area = pd.read_csv('../data/DD_AREA.dat',
                  sep='|',
                  encoding='utf-8',
                  header=None)
# 열 이름 설정
area.columns = ['Classification','SiDo_Code','SiGunGu_Code','Area_Code','SiDo_Name','SiGunGu_Name','EupMyeonDong_Name']
# 서울특별시만 뽑아서 시군구명 확인
area_seoul = area[area['SiDo_Name'] == '서울특별시']
# 시군구명을 행이름으로 바꿈
gu_st_seoul_pivot.index = area_seoul['SiGunGu_Name'].unique()
# 구별로 특정되지 않은 서울특별시 행 삭제
gu_st_seoul_pivot = gu_st_seoul_pivot.drop('서울특별시')
# 합병 위해 동일한 인덱스 이름 설정
gu_st_seoul_pivot.index.name = 'city'
# 구별 확진자수와 구별 대중교통 이용인원수 사이의 상관관계 파악을 위해 데이터 병합
gu_covid = pd.merge(covid_seoul,gu_st_seoul_pivot,on= 'city')
```

```python
# 구별 확진자 수와 구별 대중교통 인프라(정류장 수) 상관관계 파악
# 정류장 수 파악 위한 데이터 불러오기
data = pd.read_csv('../data/DM_STTN_T.dat',
                  sep='|',
                  encoding='utf-8',
                  header=None)
# 불필요한 열 삭제
for col in data.columns:
    if col==1 or col==2 or col==4 or col==6 or col==9:
        del data[col]
# 열 이름 변경
data.columns=['Operation_Date','Transportation','Station','SiDo_Code','SiGunGu_Code','Number_of_Routes']
# 데이터 중간 저장 (data_1)
data_1 = data
# 서울시 구별 정류장 수 파악을 위하여
# 서울 이외 지역을 삭제하고자 함
# 시군구 코드를 시군구 이름으로 치환해 시군구를 파악
# 시군구 이름별 코드를 알기 위해 DD_AREA.dat. 파일 재활용
# 불필요한 열 삭제
for col in area:
    if col==0 or col==3 or col==6:
        del area[col]
# 중복된 행 삭제
area_1 = area.drop_duplicates('SiGunGu_Name',keep='first')
# index 값을 '시도명'으로 치환
area_1.set_index('SiDo_Name', inplace=True)
# index가 '서울특별시'인 행만 추출
area_1 = area_1.loc['서울특별시']
# 시도코드 11이 서울특별시이므로
# data_1 데이터프레임에서 '시도코드' 값이 11인 행만 추출
data_1.set_index('SiDo_Code', inplace=True)
data_1 = data_1.loc[11]
# '정류장' 열 기준 중복 행 삭제
data_seoul = data_1.drop_duplicates('Station',keep='first')
# 운행일자, 노선수 열 삭제
del data_seoul['Operation_Date']
del data_seoul['Number_of_Routes']
# 1을 값으로 하는 열 추가
data_seoul['count']=1
# '시군구코드' 기준으로 피벗테이블 생성
data_pivot = data_seoul.pivot_table(index='SiGunGu_Code', columns='Transportation', aggfunc=sum)
#시군구명으로 인덱스 변경
data_pivot.index = area_1['SiGunGu_Name']
#서울특별시행 삭제 
data_pivot = data_pivot.drop('서울특별시')
#데이터 병합을 위해 인덱스명 통일
data_pivot.index.name = 'city'
```

```python
# 데이터 합병
gu_covid = pd.merge(gu_covid,data_pivot,on= 'city')
# 열이름 변경
gu_covid.columns = ['Confirmed','Using_Public_Transfortation','Bus_Stations','Subway_Stations']
# 그래프 그릴 데이터프레임 확인
gu_covid.head()
```

```python
# 서울시 구별 코로나 확진자 수를 월별로 정렬
# '서울' 지역만 추출
covid_info_seoul = covid_info[covid_info["province"] == '서울']
covid_info_seoul.head() 
# 특정 날짜 추출
covid_info_seoul['confirmed_month'] = [eachDate.split('-')[1] for eachDate in covid_info_seoul['confirmed_date']]
covid_info_seoul.head(10)
#서울 구별 월별 코로나 확진자수 count
covid_info_seoul.groupby([covid_info_seoul['city'],covid_info_seoul['confirmed_month']]).count()
# 코로나 확진자 수 구별 / 월별 count
covid_info_seoul = covid_info_seoul.groupby([covid_info_seoul['city'],covid_info_seoul['confirmed_month']]).count().iloc[:,0:1]
# 
covid_info_seoul = covid_info_seoul.pivot_table('patient_id',index=['city'],columns=['confirmed_month'])
# 결측치 제거
covid_info_seoul = covid_info_seoul.fillna(0)
# 정수형 변환
covid_info_seoul = covid_info_seoul.astype(int)
# 월별 확진자수 합계 열 추가
covid_info_seoul.loc[:,'sum'] = covid_info_seoul.sum(axis=1)
covid_info_seoul
# 합계열 오름차순 정렬
covid_info_seoul.sort_values(by='sum')
```

```python
#네이버 뉴스 50개 페이지 각 본문 추출 후 txt 파일로 저장
ctg_json = 'webkr.json?'
defaultURL='https://openapi.naver.com/v1/search/'
keyword = '코로나 대중교통'
search_n = 50
sort = 'sort=sim'
start = '&start=1'
display = '&display=' + str(search_n)
query = '&query=' + keyword
fullURL = defaultURL + ctg_json + sort + start + display + query
result = requests.get(urlparse(fullURL).geturl(),
         headers={"X-Naver-Client-Id":"",
                 "X-Naver-Client-Secret":""})
json_obj = result.json()
f = open("../data/news.txt", 'w', encoding="utf-8")
def get_from_bs_obj(bs_obj):
    str2=""
    for p in bs_obj.find_all('p'):
        str2 += p.text
    f.write(str2)
for item in json_obj['items']:
    try : 
        result = requests.get(item['link'])
    except Exception:
        pass
    bs_obj = BeautifulSoup(result.content, "html.parser")
    get_from_bs_obj(bs_obj)
f.close()
```

## 4. 제언 - 대체 교통수단 공공앱 '비상구' 제안

### 앱 기능 소개

대체 교통수단 공공앱 '비상구'의 기능은 다음과 같습니다.

1. 대중교통의 실시간 혼잡도 안내
    - 역 및 정류장의 실시간 인원수와 혼잡도를 알려줍니다.
    - 역 및 정류장의 대중교통 내 실시간 혼잡도를 안내하고, 혼잡도가 가장 낮은 경로를 추천합니다.
    - 지하철의 경우 탑승 예정 열차 내 혼잡도가 가장 낮은 칸을 안내합니다.
    - 버스의 경우 혼잡도가 상대적으로 낮은 대체 버스를 제시합니다.

<div>
</div>

2. 대중교통 대체 교통수단(따릉이, 전기자전거, 전동스쿠터 등)을 활용한 경로 안내
    - 사용자가 출발지와 목적지를 입력하면, 자동으로 혼잡도가 낮은 경로를 추천합니다.
    - 이 때 혼잡한 대중교통 정류장 인근에 대체 교통수단 대여소가 있다면, 이를 활용한 경로로 안내합니다.
    - 앱 내에는 특정 대여소의 잔여 대체 교통수단이 표시되어 이용자의 선택을 돕습니다.
    - 또한 성별, 연령대별 맞춤 주행 경로를 추천함으로써 이용자의 편의를 증대시킵니다.

이러한 공공앱 '비상구'를 통해 대중교통 내 혼잡도를 낮추는 효과를 기대할 수 있습니다. 이는 사용자 개인과 대중교통 이용자의 코로나 감염 확률을 낮출뿐더러, 대중교통 혼잡도를 낮춰 사람들의 인식을 개선하는데 도움을 줄 수 있습니다.

### > 성별, 연령대별 맞춤 주행거리 안내

```python
#따릉이 성별 연령별 평균이동거리
bicycle = pd.read_csv('../data/공공자전거 이용정보(일별)_2020.06.csv', encoding= 'ANSI')
# 총 데이터에서 필요한 열만 추출하기
bicycle = bicycle[['성별','연령대코드','이동거리(M)']]
bicycle.columns = ['Sex','Age','Movement_distance(M)']
#성별 열에서 '\N'값 제외하고 추출하기
idx = bicycle[bicycle['Sex']==r'\N'].index
bicycle.drop(idx, inplace=True)
# m,f,M,F 로 분리된 데이터 대문자로 정렬하기
bicycle['Sex'] = bicycle['Sex'].str.upper()
# M -> 남성 으로 데이터명 변경
bicycle = bicycle.replace('M','Male')
# F -> 여성 으로 데이터명 변경
bicycle = bicycle.replace('F', 'Female')
# men = bicycle['성별']=='M'
# age10 = bicycle['연령대코드']=='AGE_001'
# men_age10 = bicycle[men & age10]
# men_age10.mean()
# 연령대, 성별 기준으로 평균이동거리 groupby 정렬
bicycle_group = bicycle.groupby(['Age','Sex'])[['Movement_distance(M)']].mean()
# int형태 데이터프레임으로 추출
bicycle_group.astype(int).unstack()
```
