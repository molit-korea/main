- 저희 팀은 익히 알려진 정보를 검증하는 것부터 시작하여, 그 중 새로운 발견을 하고 다시 그 것에 대해서 검증하는 것을 반복하는 식의 Bottom-Up approach를 시도했습니다.  
- 또한, 분석 내내 <u>'현재 이슈가 되는 것과 가까운 미래에 대한 실용적인 결과'</u>에 항상 초점을 맞추었습니다.

```python
### 기사 크롤링 결과와 시각화에 필요한 폰트 및 마스크를 다운로드합니다 
!wget --no-check-certificate 'https://docs.google.com/uc?export=download&id=11Otqnc2OV6uS-_orSOccqJBO_tSATLS4' -O merged_news.zip
!unzip merged_news.zip
!wget --no-check-certificate 'https://docs.google.com/uc?export=download&id=1YKS0815cGD20YK3IcMjiSpBBK1ezsd1-' -O BMJUA_ttf.ttf
!wget --no-check-certificate 'https://docs.google.com/uc?export=download&id=1IRc2G-GFwOT-l7AQoZQTZslYwcjbkHc3' -O jammanbo_mask.png
```

```python
target_base_path = 'merged_news'

nouns_cnt_list = []
for month_i in range(8):
    start_date = '2020{:02d}01'.format(month_i+1)
    end_date = '2020{:02d}01'.format(month_i+2)
    target_filename = pth.join(target_base_path, '{}_{}_{}'.format(start_date, end_date, 'merged_news.txt'))
    with open(target_filename, encoding='cp949') as f:
        text = f.read()

    nouns = []
    api = KhaiiiApi()
    for word in api.analyze(text):
        for morph in word.morphs:
            if morph.tag in ['NNG', 'NNP'] and len(morph.lex) > 1:
                nouns.append(morph.lex)


    # stop_words = []
    stop_words = [
        '신종', '코로나바이러스', '코로나', '코로나19', '바이러스', 
        '확인', '코로', '이날', '경우', '이후', '때문', '지난달'
        '가능', '가운데', '강조', '관련', '로나', '사실', '이번', '올해',  
        '전날', '정도', '지금', '지난해', '처음', '최근', '포함', '하루',  
    ]
    nouns = [each_word for each_word in nouns if each_word not in stop_words]

    cnt = Counter(nouns).most_common(200)
    dic_cnt = dict(cnt)
    
    nouns_cnt_list.append(dic_cnt)

    
def color_func(word, font_size, position, orientation, random_state=None,
                    **kwargs):
    return "hsl(0, %d%%, %d%%)" %(random.randint(0, 10), random.randint(90, 100))    
    
nanum_font_base_path = '/usr/share/fonts/truetype/nanum'
nanum_font_path = pth.join(nanum_font_base_path, 'NanumSquareRoundR.ttf')

font_dirs = ['/usr/share/fonts/truetype/nanum/', './']
font_files = fm.findSystemFonts(fontpaths=font_dirs)
font_list = fm.createFontList(font_files)
fm.fontManager.ttflist.extend(font_list)

font_name = fm.FontProperties(fname=nanum_font_path).get_name()
plt.rc('font', family=font_name)
plt.rcParams.update({'font.size': 15})
mpl.rcParams['axes.unicode_minus'] = False

wc_font_path = 'BMJUA_ttf.ttf'

result_base_path = 'wc_result'
os.makedirs(result_base_path, exist_ok=True)

wc_list = []
for month_i in range(8):
    dic_cnt = nouns_cnt_list[month_i]
    
    MASK = np.array(Image.open(pth.join("jammanbo_mask.png")))
    wc = WordCloud(
        font_path=wc_font_path, 
        background_color="black",
        max_words=200,
        mask=MASK,
        width=1000, 
        height=1000,
        color_func=color_func,
    )
    wc.generate_from_frequencies(dic_cnt)
    result_filename = pth.join(result_base_path, '{}_{}_{}'.format(start_date, end_date, 'result.png'))
    wc.to_file(result_filename)
    wc_list.append(wc)        

    
# WordCloud
plt.figure(figsize=(15, 7))
for month_i in range(8):
    plt.subplot(2,4,(month_i+1))
    plt.imshow(wc_list[month_i], interpolation='bilinear')
    plt.xticks([])
    plt.yticks([])
    plt.title('{:02d}월'.format(month_i+1))
plt.suptitle('코로나로 검색한 기사 월별 빈도분석 결과')
plt.show()
print()

# Rank plot
def get_rank_dict(target_nouns_cnt_list):
    temp_cnt_list = sorted(list(target_nouns_cnt_list.items()), key=lambda v: v[1], reverse=True)
    nouns_rank_dict = {each_keyword:i for i, (each_keyword, each_cnt) in enumerate(temp_cnt_list)}
    return nouns_rank_dict

plt.figure(figsize=(9,3))
# plt.box(False)
for target_keyword in ['확진자', '미국', '중국', '서울']:
    target_rank = [get_rank_dict(nouns_cnt_list[month_i])[target_keyword] for month_i in range(8)]
    sns.lineplot(np.arange(len(target_rank)), target_rank, marker='o', label=target_keyword)
plt.xlim(-0.5, 9)
plt.ylim(-10, 180)
plt.xticks(np.arange(8), list(map(lambda x: str(x)+'월',np.arange(8)+1)))
plt.ylabel('순위')
plt.gca().invert_yaxis()
plt.legend(loc='upper right', fontsize=12)
plt.title('기간에 따른 일부 키워드의 순위 변동성')
plt.show()
```

## 3.4. 연령대 별 유동인구와 확진자의 상관관계 분석

#### 접근법
- [2단계 1.C.d.] 연령대 별 유동인구와 확진자의 관계를 알아보기 위해서 각 연령별 비율을 시각화하고 피어슨 상관계수(Pearson's Correlation Coefficient)를 표시하였습니다.
- 확진자가 처음 발생한 1월 20일부터 데이터의 마지막 날짜인 6월 30일까지의 데이터만 사용하였습니다.<br>KT와 SKT 유동인구 데이터 각각의 양과 연령범위가 달라서 연령 별 유동인구 백분율로 바꾼 후 평균을 내는 식으로 전처리하였습니다.
- 나이대는 0~70대, 10살 간격으로 동일하게 맞추었습니다.

#### 결과
1. [2단계 1.C.d. 나머지 검증] <u>3\~40대 유동인구가 가장 많으며, 2\~50대 유동인구가 전체의 80% 가까이 된다</u>는 것을 확인할 수 있었습니다.<br>
2. 계산된 피어슨 상관계수값을 보면 <u>연령대 별 유동인구와 코로나 확진자의 상관성이 놀라울 정도로 높다</u>는 것을 확인할 수 있었는데, <u>저연령(0\~10대)대와 고연령대(50\~70대)의 추이가 비슷한 것이 그 원인</u>인 듯 합니다.

## 3.6. Granger Causality를 통한 확진자 정보와 유동인구 변화의 인과관계 분석

```python
def multipolygon_to_coordinates(x):
    if x.type == 'Polygon':
        lon, lat = x.exterior.xy
        return [[x, y] for x, y in zip(lon, lat)]
    else:
        result_list = []
        for each_x in x:
            lon, lat = each_x.exterior.xy
            for x_, y_ in zip(lon, lat):
                result_list.append([x_, y_])
        return result_list
```

- 앞서 언급했듯이, 버스나 지하철에서의 유동인구는 출퇴근인원이 있는 이상 어쩔 수 없이 발생할 수 밖에 없기에 <u>방역</u>과 같은 대책이 중요합니다.
- 꾸준히 하고 있다고는 하나, 현재 지하철이나 버스에서의 방역은 사진과 같이 종점에서 사람들이 다 내렸을 때 행해지고 있는 상황입니다.
- 실상 사람들이 많이 부대낄 때 전파가 일어나는 것을 생각하면, <u>사람들이 없을 때가 아니라 사람들이 많을 때 더 방역을 해야 한다</u>고 할 수 있습니다.
- 하지만 20대 인력을 투입한다고 해도 인력의 한계라는 것이 있으니, 그 <u>방역인력분배를 저희 팀이 계산한 시간대 별 유동인구 지표를 활용해서 한다면 훨씬 효율적으로 방역을 할 수 있을 것</u>입니다.
- 가장 중요한 것은 <u>현실성</u>입니다. <br>저희 팀의 분석 결과에 따라, <u>유동인구 수가 집중되는 특정 시간대에만 아르바이트 형식으로 인력을 고용한다면 저렴한 비용으로 큰 효과</u>를 낼 수 있을 것입니다.
- 또한 방역에는 특별한 기술이나 힘이 필요하지 않기에 <u>빠르게 새 인력을 투입하고 효과를 볼 수 있다는 점</u>에서 더 현실적입니다.
- 저희 팀에 분석 결과를 보면 고령 인구에 대한 방역 또한 시급한 것을 알 수 있습니다.<br> 분석 결과에 따라 <u>고령화 인구가 밀집한 지역에 방역을 조금 더 신경쓴다면 고령 확진자 수를 줄일 수 있을 것</u>입니다.

- 마스크를 착용하지 않은 사람이 있었을 때 일반적으로 유동인구가 높은 쪽이 더 위험할 것입니다.<br>
<u>마스크 단속을 하는 경우에도 저희 팀의 교통 분석 데이터는 유용하게 사용될 수 있을 듯 합니다</u>

## 5.4. 제안 3:명절맞이 유동인구 체크 인력 고용 


<img src="https://img-s-msn-com.akamaized.net/tenant/amp/entityid/BB12cWth.img?h=0&w=600&m=6&q=60&u=t&o=f&l=f" style="height:200px;"/>

- 귀향길에 오르는 서울 거주민들 덕에 고속도로나 KTX, 새마을호 등은 명절마다 북적입니다.
- 저희 팀의 분석 결과에서도 명절에는 평소보다 기차 탑승객이 증가한다는 것이 검증되었습니다.
- [4.5.]에서 언급했듯이, 전국적으로 인구가 섞이는 이 시기에 코로나 환자가 섞여있다가는 자칫 아비규환이 벌어질 수 있습니다.

<img src="https://dimg.donga.com/wps/NEWS/IMAGE/2020/06/12/101486568.1.jpg" style="height:200px;"/>


- 요즘은 어디를 가든 출입 기록과 체온을 적고 출입하게 되어있습니다.
- 유동인구가 가장 많이 발생하는 대중교통은 이동 경로는 할 수 있을 지 몰라도 제일 중요한 체온을 체크할 수 없다는 맹점이 있습니다.
- 현재 대중교통 인력으로 이를 체크하는 것은 불가능에 가깝습니다. 
- 따라서 저희 팀은 저희 팀은 <u>대중교통 유동인구 체크 아르바이트</u>를 제안합니다

- 이 제안의 장점은 명절이 아니라도 꼭 필요한 인력이며, 대중교통은 전국적으로 고르게 퍼져있기 때문에 <u>일자리도 고르게 분포되어 있다</u>는 점입니다.
- 또한 [5.3과 마찬가지로] 체온 체크에는 별다른 기술이나 자격이 필요 없기 때문에, <u>빠르게 새 인력을 투입하고 효과를 볼 수 있다는 점</u>과 <u>유동인구 수가 집중되는 특정 시간대에만 인력을 투입한다면 저렴한 비용으로 큰 효과</u>를 볼 수 있다는 현실적인 점 또한 동일합니다.
