# 결과


2019년에 비하여 2020년의 지하철 및 여객 이용량이 급격히 감소하였으며, 그 감소요인은 코로나 국내확산에 의한 것으로 확인되었다. <br>
코로나 확산이 국민들의 전체 이동량에 영향을 주는 것인지 확인하기 위하여 인구가 밀집해있는 수도권을 중심으로 교통량을 확인했으며,<br>
그 결과 첫 확진자가 발생하면서 약간의 교통량 감소가 있었으며, 2월 18일 대구에서 대규모 집단감염이 발생한 후 재택근무를 실시하게 되면서 교통량이 큰 폭으로 감소하였다.<br>
3, 4월에 코로나 유행이 점차 가라앉으며 교통량이 다시 증가하게 되었다. 4월 말~ 5월 초 황금연휴 기간 이후 생활 속 거리 두기로 전환됨에 따라 수도권에서 지역 내 집단감염이 발생하여 교통량이 소폭 감소하였다.<br><br>


지하철 전체적인 이용객 수를 살펴보면 코로나 확산에 의해 이용객 수가 급격히 하락한 것을 알 수 있다.<br>
먼저 1월 25일 이후 우한에서 코로나 바이러스가 유행하기 시작했을 때 전체 평균 9% 정도로 소폭 감소하였고, 국내에서 확진자가 점차 늘어나기 시작한 2월 18일 이후에는 평균 20% 정도로 대폭 감소하였다.<br>
그 이후에는 Epidemic Curve의 하락하는 지점 (대략 3/1)부터 다시 천천히 이용객 수가 회복하는 것으로 보인다.<br><br>


지하철 이용객 수의 감소의 가장 큰 이유는 수도권 전체인구의 이동량이 줄었고, 재택근무로 인하여 출퇴근하는 인원이 줄어든 영향이 크기 때문이다.  <br>
두 가지 이유로 꼽을 수 있었던 가장 큰 단서는 일주일 단위로 반복되는 이용량 패턴으로서 찾을 수 있었다.<br>
주말과 공휴일에는 급격히 떨어지고, 월요일에 급격히 증가하며 일부지만 수요일 혹은 목요일에 살짝 감소하며, 금요일에는 뾰족한 산을 보이고 주말이 돌아오면 다시 감소한다. <br>
이로써 이용객의 이동 패턴의 변화가 아닌 단순하게 이용객이 감소했다고 유추할 수 있었다.<br><br>


특히 서울역과 용산역은 직장뿐만 아니라 기차를 타는 사람들이 많이 찾는 역으로 다른 역들에 비해 감소 폭이 매우 컸고, 3, 4월이 지났을 때도 다시 돌아오는 상승세가 매우 약하다.<br>
특히 공항철도 이용객은 코로나가 잠식되기 전까지는 회복이 어려울 것으로 보인다. <br><br>

하지만 머지않아 다른 역들처럼 다시 회복할 수 있다고 가정하면 전체 지하철 이용객은 다시 늘어날 것이다.<br>
그럴 때 다른 교통수단에 비해 더욱 폐쇄적인 구조를 갖는 지하철은 잠재적 위험구역이 아닌 실제 화약고가 될 수 있는 조건을 갖고 있다. <br>

따라서 열차 도착 시간 간격을 좁히거나 원래의 배차 간격을 유지하여 승객 간의 안전거리 확보가 될 수 있게 하는 것이 필요하다.<br><br>

````python
patient = pd.read_csv('../data/PatientInfo.csv')
sbwy2019 = pd.read_csv('../data/KR_TB_SBWY_STN_HR_CON_20190101_20190531_UTF8.csv')
sbwy2020 = pd.read_csv('../data/KR_TB_SBWY_STN_HR_CON_20200101_20200531_UTF8.csv')
````

### 1.1.2 코로나 확진자 자료 정리

````python
patient['confirmed_date'] = pd.to_datetime(patient['confirmed_date'], format = '%Y-%m-%d')

patient = patient.set_index('confirmed_date')

confirmed = patient.resample('1D').count()
confirmed = confirmed['patient_id']
confirmed = pd.DataFrame(confirmed)
confirmed.columns = ['count']

confirmed.reset_index(drop = False, inplace = True)

confirmed.head()
````

````python
def datetime_to_monthday(x):
    x = str(x)
    x = x.replace(" ", "")
    x = x.replace(",", "")
    x = x.replace("-", "")
    x = x.replace(":", "")
    return(x[ 4:8 ]) 

confirmed['confirmed_date'] = confirmed['confirmed_date'].apply(datetime_to_monthday)
confirmed
````

### 1.1.3 subway 자료 정리 - 지하철

````python
sbwy2019.columns = ['Date', 'stn_no', 'stn_name', 'time', 'abrd', 'goff']
sbwy2020.columns = ['Date', 'stn_no', 'stn_name', 'time', 'abrd', 'goff']
````

````python
def space_comma_remove(x):
    x = x.replace(" ", "")
    x = x.replace(",", "")
    return x

sbwy2019['abrd'] = sbwy2019['abrd'].apply(space_comma_remove)
sbwy2019['goff'] = sbwy2019['goff'].apply(space_comma_remove)

sbwy2020['abrd'] = sbwy2020['abrd'].apply(space_comma_remove)
sbwy2020['goff'] = sbwy2020['goff'].apply(space_comma_remove)
````

````python
sbwy2019['abrd'] = np.where(sbwy2019['abrd'] == '-', 0, sbwy2019['abrd'])
sbwy2019['goff'] = np.where(sbwy2019['goff'] == '-', 0, sbwy2019['goff'])

sbwy2020['abrd'] = np.where(sbwy2020['abrd'] == '-', 0, sbwy2020['abrd'])
sbwy2020['goff'] = np.where(sbwy2020['goff'] == '-', 0, sbwy2020['goff'])
````

````python
def error_to_zero(a) :
    try:
        my_value = int(a)
    except TypeError:
        my_value = 0
    return my_value

sbwy2019['abrd'] = sbwy2019['abrd'].apply(error_to_zero)
sbwy2019['goff'] = sbwy2019['goff'].apply(error_to_zero)

sbwy2020['abrd'] = sbwy2020['abrd'].apply(error_to_zero)
sbwy2020['goff'] = sbwy2020['goff'].apply(error_to_zero)
````

````python
sbwy2019 = sbwy2019.astype({'Date' : str, 'stn_no' : str})
sbwy2020 = sbwy2020.astype({'Date' : str, 'stn_no' : str})

sbwy_sum_2019 = sbwy2019.groupby(sbwy2019['Date']).sum()
sbwy_sum_2020 = sbwy2020.groupby(sbwy2020['Date']).sum()

sbwy_sum_2019["total"] = sbwy_sum_2019['abrd'] + sbwy_sum_2019['goff']
sbwy_sum_2020["total"] = sbwy_sum_2020['abrd'] + sbwy_sum_2020['goff']

del sbwy_sum_2019['abrd']
del sbwy_sum_2019['goff']

del sbwy_sum_2020['abrd']
del sbwy_sum_2020['goff']

sbwy_sum_2019.reset_index(drop = False, inplace = True)
sbwy_sum_2020.reset_index(drop = False, inplace = True)

sbwy_sum_2019['Date'] = sbwy_sum_2019['Date'].apply(datetime_to_monthday)
sbwy_sum_2020['Date'] = sbwy_sum_2020['Date'].apply(datetime_to_monthday)

sbwy_sum_2019
sbwy_sum_2020
````

### 1.1.5 Tableau 자료 만들기 및 저장 - 지하철

````python
confirmed.columns = ['Date', 'count']
````

````python
sbwy_2019_2020 = pd.merge(sbwy_sum_2019, sbwy_sum_2020, how = 'inner', on = 'Date')
sbwy_2019_2020
````

````python
sbwy_2019_2020_confirmed = pd.merge(sbwy_2019_2020, confirmed, how = 'outer', on = 'Date')

sbwy_2019_2020_confirmed.columns = [ 'date', '2019', '2020', 'corona' ]

def con_2020( x ) :
    x = "2020" + x
    return x

sbwy_2019_2020_confirmed['date'] = sbwy_2019_2020_confirmed['date'].apply(con_2020)
sbwy_2019_2020_confirmed
````

````python
def gather( df, key, value, cols ) :
    id_vars = [ col for col in df.columns if col not in cols ]
    id_values = cols
    var_name = key
    value_name = value
    return pd.melt( df, id_vars, id_values, var_name, value_name )
    
sbwy_2019_2020_confirmed = gather( sbwy_2019_2020_confirmed, "year", "count", [ '2019', '2020', 'corona' ] )
````

그럴 때 다른 교통수단에 비해 더욱 폐쇄적인 구조를 갖는 지하철은 잠재적 위험구역이 아닌 실제 화약고가 될 수 있는 조건을 갖고 있다.

#### 1.4.1.3 자료 병합

````python
trn_2020 = trn_2020.merge(tr_stn_province, how = "left", on = 'stn_name')
trn_2020 = trn_2020[~trn_2020['province'].isnull()].reset_index(drop = True)
````

#### 1.4.1.4 승.하차 자료 타입 숫자로 변형하기

````python
def space_remove(x):
    x = x.replace(" ", "")
    return x

def comma_remove(x):
    x = x.replace(",", "")
    return x
````

````python
trn_2020['abrd'] = trn_2020['abrd'].apply(space_remove)
trn_2020['goff'] = trn_2020['goff'].apply(space_remove)

trn_2020['abrd'] = trn_2020['abrd'].apply(comma_remove)
trn_2020['goff'] = trn_2020['goff'].apply(comma_remove)

trn_2020['abrd'] = np.where(trn_2020['abrd'] == '-', 0, trn_2020['abrd'])
trn_2020['goff'] = np.where(trn_2020['goff'] == '-', 0, trn_2020['goff'])
````

````python
def error_to_zero(x) :
    try:
        my_value = int(x)
    except TypeError:
        my_value = 0
    return my_value
````

````python
trn_2020['abrd'] = trn_2020['abrd'].apply(error_to_zero)
trn_2020['goff'] = trn_2020['goff'].apply(error_to_zero)
````

````python
trn_2020 = trn_2020.astype({'Date' : str, 'stn_no' : str})
trn_2020 = trn_2020.groupby([trn_2020['Date'], trn_2020['province']]).sum()
trn_2020.reset_index(drop = False, inplace = True)

trn_2020.head()
````

<a name="Random_Forest"></a>

## 2.3 Random Forest

### 2.3.1 평가 지표

````python
def MAE(y_true, y_pred): 
    return np.mean(np.abs((y_true - y_pred)))
````

### 2.3.2 Model Tuning - Bayesian Optimization

#### 2.3.2.1 Random Foreset Parameter setting

````python
rf_param_bounds = {'n_estimators': (100, 1000), # 결정 트리의 개수
                   'max_depth': (3, 10), # 트리의 깊이
                   'min_samples_leaf': (3, 10), # 리프노드가 되기 위한 최소한의 샘플 데이터 수
                   'min_samples_split': (2, 10) # 노드를 분할하기 위한 최소한의 데이터 수
                  }
````

#### 2.3.2.2 Bayesian Optimization 적용을 위한 함수 구현

````python
def rf_hyper_param(n_estimators, max_depth, min_samples_leaf, min_samples_split, df = df, n_splits = 10):
    
    cv = KFold(n_splits = n_splits, shuffle = True, random_state = 123)
    
    mae_res = []
    
    for i, (idx_train, idx_test) in enumerate(cv.split(df)) :
        df_train = df.iloc[idx_train]
        df_test = df.iloc[idx_test]
        
        df_train_X = df_train.drop('target', axis=1)
        df_train_y = df_train.target

        df_test_X = df_test.drop('target', axis=1)
        df_test_y = df_test.target
    
        n_estimators = int(n_estimators)
        max_depth = int(max_depth)
        min_samples_leaf = int(min_samples_leaf)
        min_samples_split = int(min_samples_split)
    
        clf = RandomForestRegressor(n_estimators = n_estimators, max_depth = max_depth, min_samples_leaf = min_samples_leaf, min_samples_split = min_samples_split, random_state = 123)
        clf.fit(df_train_X, df_train_y)

        rf_pred_after_optim = clf.predict(df_test_X).round(0)
        temp = MAE(df_test_y, rf_pred_after_optim)
    
        mae_res.append(temp)
    
    return -np.mean(mae_res)
````

#### 2.3.2.3 Bayesian Optimization 적용

````python
rf_optimizer = BayesianOptimization(f = rf_hyper_param, pbounds = rf_param_bounds, random_state = 123)
rf_optimizer.maximize(init_points = 5, n_iter = 50)
````

### 2.3.3 최종 Paramter 선정 후 모델 적용

````python
rf_bayes_params = rf_optimizer.max['params']

Bayes_RF = RandomForestRegressor(n_estimators = int(rf_bayes_params['n_estimators']), 
                                 max_depth = int(rf_bayes_params['max_depth']), 
                                 min_samples_leaf = int(rf_bayes_params['min_samples_leaf']), 
                                 min_samples_split = int(rf_bayes_params['min_samples_split']),
                                 random_state = 123)

Bayes_RF.fit(X_train, y_train)
````

### 2.3.4 Test 데이터에 대한 MAE 확인

````python
rf_pred_after_optim = Bayes_RF.predict(X_test).round(0)
rf_mae_after_optim = MAE(y_test, rf_pred_after_optim)

print("Bayesian optimization MAE : %.4f" % rf_mae_after_optim)
````

### 2.3.5 모델 저장 및 결과 값 저장

## 2.7 CatBoost

### 2.7.1 평가 지표

````python
def MAE(y_true, y_pred): 
    return np.mean(np.abs((y_true - y_pred)))
````

### 2.7.2 Model Tuning - Bayesian Optimization

#### 2.7.2.1 CatBoost Parameter setting

````python
cat_param_bounds = {'learning_rate' : (0.001, 0.1),
                    'iterations':  (100, 1000),
                    'depth': (2, 10),
                    'subsample': (0.4, 1),
                    'l2_leaf_reg': (0, 1)
                    }
````

#### 2.7.2.2 Bayesian Optimization 적용을 위한 함수 구현

````python
def cat_hyper_param(learning_rate, iterations, depth, subsample, l2_leaf_reg, df = df, n_splits = 10):
    
    cv = KFold(n_splits = n_splits, shuffle = True, random_state = 123)
    
    mae_res = []
    
    for i, (idx_train, idx_test) in enumerate(cv.split(df)) :
        df_train = df.iloc[idx_train]
        df_test = df.iloc[idx_test]
        
        df_train_X = df_train.drop('target', axis=1)
        df_train_y = df_train.target

        df_test_X = df_test.drop('target', axis=1)
        df_test_y = df_test.target
    
        iterations = int(iterations)
        depth = int(depth)

        clf = catboost.CatBoostRegressor(learning_rate = learning_rate, iterations = iterations, 
                                         depth = depth, subsample = subsample, l2_leaf_reg = l2_leaf_reg, thread_count = 7)

        clf.fit(df_train_X, df_train_y, eval_set=(df_test_X, df_test_y), use_best_model = True, verbose = False)

        catreg_pred_after_optim = clf.predict(df_test_X).round(0)
        temp = MAE(df_test_y, catreg_pred_after_optim)
        
        mae_res.append(temp)
    
    return -np.mean(mae_res)
````

#### 2.7.2.3 Bayesian Optimization 적용

````python
cat_optimizer = BayesianOptimization(f = cat_hyper_param, pbounds = cat_param_bounds, random_state = 123)
cat_optimizer.maximize(init_points = 5, n_iter = 50)
````

### 2.7.3 최종 Paramter 선정 후 모델 적용

````python
cat_bayes_params = cat_optimizer.max['params']

Bayes_CAT = catboost.CatBoostRegressor(learning_rate = cat_bayes_params['learning_rate'], 
                                          iterations = int(cat_bayes_params['iterations']), 
                                          depth = int(cat_bayes_params['depth']), 
                                          subsample = cat_bayes_params['subsample'], 
                                          l2_leaf_reg = cat_bayes_params['l2_leaf_reg'], 
                                          thread_count = 5)

Bayes_CAT.fit(X_train, y_train, verbose = False)
````

````python
cat_pred_after_optim = Bayes_CAT.predict(X_test).round(0)
cat_mae_after_optim = MAE(y_test, cat_pred_after_optim)

print("Bayesian optimization MAE : %.4f" % cat_mae_after_optim)
````

### 3.2.9 CatBoost

- 성능이 제일 좋은 모형으로 진행

````python
def MAE(y_true, y_pred): 
    return np.mean(np.abs((y_true - y_pred)))
````

````python
cat_param_bounds = {'learning_rate' : (0.001, 0.1),
                    'iterations':  (100, 1000),
                    'depth': (2, 10),
                    'subsample': (0.4, 1),
                    'l2_leaf_reg': (0, 1)
                    }
````

````python
def cat_hyper_param(learning_rate, iterations, depth, subsample, l2_leaf_reg, df = df, n_splits = 10):
    
    cv = KFold(n_splits = n_splits, shuffle = True, random_state = 123)
    
    mae_res = []
    
    for i, (idx_train, idx_test) in enumerate(cv.split(df)) :
        df_train = df.iloc[idx_train]
        df_test = df.iloc[idx_test]
        
        df_train_X = df_train.drop('confirmed_people', axis=1)
        df_train_y = df_train.confirmed_people

        df_test_X = df_test.drop('confirmed_people', axis=1)
        df_test_y = df_test.confirmed_people
    
        iterations = int(iterations)
        depth = int(depth)

        clf = catboost.CatBoostRegressor(learning_rate = learning_rate, iterations = iterations, 
                                         depth = depth, subsample = subsample, l2_leaf_reg = l2_leaf_reg, thread_count = 7)

        clf.fit(df_train_X, df_train_y, eval_set=(df_test_X, df_test_y), use_best_model = True, verbose = False)

        catreg_pred_after_optim = clf.predict(df_test_X).round(0)
        temp = MAE(df_test_y, catreg_pred_after_optim)
        
        mae_res.append(temp)
    
    return -np.mean(mae_res)
````

````python
cat_optimizer = BayesianOptimization(f = cat_hyper_param, pbounds = cat_param_bounds, random_state = 123)
cat_optimizer.maximize(init_points = 5, n_iter = 50)
````

````python
cat_bayes_params = cat_optimizer.max['params']

Bayes_CAT_corona = catboost.CatBoostRegressor(learning_rate = cat_bayes_params['learning_rate'], 
                                              iterations = int(cat_bayes_params['iterations']), 
                                              depth = int(cat_bayes_params['depth']), 
                                              subsample = cat_bayes_params['subsample'], 
                                              l2_leaf_reg = cat_bayes_params['l2_leaf_reg'], 
                                              thread_count = 5)

Bayes_CAT_corona.fit(X_train, y_train, verbose = False)
````

````python
cat_pred_after_optim = Bayes_CAT_corona.predict(X_test).round(0)
cat_mae_after_optim = MAE(y_test, cat_pred_after_optim)

print("Bayesian optimization MAE : %.4f" % cat_mae_after_optim)
````

## 3.3 모델에 대한 결과 해석

자료의 형태는 시간에 따라 자료의 값이 계속 바뀌는 다이내믹 데이터 마트(Dynamic data mart)로 구성되어있다. <br>
과거의 자료로 미래를 예측하는 데이터 마트를 구성해야 트리 기반 모형을 적용하였을 때에도 시계열적인 결과를 볼 수 있다. <br>


먼저, `2.8 결과 정리`에서 검증 자료의 실제 결과를 시각화 하였으며, 시간에 따라 교통량의 추세를 잘 따라감을 확인할 수 있다. <br>
**`2.8.1 평가 지표 확인`**을 통해 검증 자료를 제일 잘 예측한 모형은 CatBoost이며, 자세한 설명은 링크 부분을 보면 많은 도움이 될 것이다.  <br>
교통량을 예측하는 모델링을 통해 현재 만들어 놓은 모형이 잘 설명함을 시각적으로 확인할 수 있다. <br>


시각적으로 표현한 지역에 따른 교통량 예측 결과에 대해 각 설명변수들이 어떻게 반응하는지를 확인을 해보았으며, 코로나와 교통량의 관계를 살펴보았다. <br>
복잡한 Black-Box 모형을 해석하기 위해 Shap 방법론을 이용하였으며, 자세한 설명은 링크 부분을 보면 많은 도움이 될 것이다.
Shapley value를 통해 모델 해석을 하였을 때, 코로나가 상승하는 구간 그리고 주말에는 교통량은 감소하였고 코로나 발생과 주말은 교통량에 영향을 유의미하게 줌을 확인하였다. <br> 


다음으로 코로나 예측을 위해 교통량 자료를 활용해보았으며, Shapley value를 통해 코로나 발생에 **휴일**이 유의미한 결과를 주어 처음 제시한 결과에 뒷받침할 수 있는 근거가 됨을 확인하였다.
